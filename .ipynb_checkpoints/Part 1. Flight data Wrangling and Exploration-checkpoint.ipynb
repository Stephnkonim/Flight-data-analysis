#### We need to get data for just flights that were cancelled

```python
#To get just data for just flights that were cancelled.
Cancelled_flights = Flight_data.loc[Flight_data['CancellationCode'].isin(['A','B','C','D'])]
Cancelled_flights.head(5)
```

```python
# using a barchart to show the distinct count of each code.
sb.countplot(data=Cancelled_flights, x='CancellationCode', color='darkred', order=['A','B','C','D']);
plt.title('Frequency of each cancellation cause')
plt.show();
```

```python
#plotting a pie chart to further expand on our investigation
sorted_counts = Cancelled_flights['CancellationCode'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False);
print(sorted_counts)
plt.title('Frequency of each cancellation cause')
plt.axis('square');
```

From the bar chart  and pie chart, it is clear that the most likely reason for a flight to be cancelled is the state of the carrier, the weather is next most likely factor.However for the first 4 months of 2008, security reasons have not affected flight cancellations in a significant way.

> Concerning Flights that were delayed. I'll like to look at features of these these flights to check for any significant information

> I need to get data for delayed flights only. To do this I will first remove all data when the departure delay time is 0, and then drop all rows where any delay feature had a null value

>Also all rows with departure delay time as negative values will also be dropped as negative values mean the flight was earlier than scheduled

#### We need to extract data for only delayed flights. however since we are interested in the features, we will also remove delayed flights that do not have values for any delay feature.

```python
#changing all 0 values to Null 
columns = ['ArrDelay','DepDelay','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']

Flight_data[columns] = Flight_data[columns].replace(['None', 0], np.nan)
```

```python
#Melting the 4 feature columns into just 2, Delfeature and Delfeature_value
Delayed_flight_data = pd.melt(Flight_data,id_vars=['Year','Month','DayofMonth','DayOfWeek','DepTime','CRSDepTime','ArrTime','CRSArrTime','UniqueCarrier',
                                                        'FlightNum','TailNum','ActualElapsedTime','CRSElapsedTime','AirTime','ArrDelay','DepDelay','Origin','Dest','Distance'], 
                              value_vars =['CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay'],
                              var_name = 'Delfeature' ,
                              value_name = 'Delfeature_time')
```

```python
#Drop all records with Nan values in the Delfeature_value columns
Delayed_flights = Delayed_flight_data.dropna(subset=['Delfeature_time','DepDelay'] )
Delayed_flights.head(5)
```

```python
Delayed_flights.shape
```

```python
Delayed_flights['DepDelay'].min()
```

```python
#to remove negative flight delay times
Delayed_flights=Delayed_flights.drop(Delayed_flights.index[Delayed_flights['DepDelay'] < 0])
```

```python
Delayed_flights['DepDelay'].min()
```

```python
Delayed_flights.describe()
```

Delay time has a lot of points  with low vlaues, with a very long tail od data pointa with large alues, with the logarithmic tranform. using a log-scale, the peak of our distribution is between 30  to 60, Also Delay times beyond 600 minutes (10hrs) are negligible and will be considered as outliers

##### I'll like to see the change over time of the number of cancelled flights during the first 4 months

```python
Cancelled_flights.head()
```

```python
Cancelled_flights_with_date = Cancelled_flights.copy()
```

```python
#Change data type of year, Month and day of the month to string. 
Cancelled_flights_with_date[['Year','Month','DayofMonth','DayOfWeek']] = Cancelled_flights_with_date[['Year','Month','DayofMonth','DayOfWeek']].astype(str)

#Change data type of year, Month and day of the month to string. 
Cancelled_flights_with_date['Date']=Cancelled_flights_with_date["Year"]+ '-' +Cancelled_flights_with_date['Month']
```

```python
#convert the date column to date time
Cancelled_flights_with_date['Date']= pd.to_datetime(Cancelled_flights_with_date['Date'])
```

```python
Cancelled_flights_with_date=Cancelled_flights_with_date[['Date','Cancelled']]
```

```python
#grouping by count of delay time
C= Cancelled_flights_with_date.groupby('Date')['Cancelled'].count()
```

```python
C. head()
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(12,10))
plt.xlabel("Date")
plt.ylabel('Number of cancelled flights')
plt.title("Cancelled fights over the first four months")
plt.xticks(rotation=90);
plt.ylim(1000,25000,1000)
plt.plot(C)
```

From the graph, over 15000 flights were cancelled in January, the numbers increased in february to over 20000 flights, it gradually reduces to between  15,000 and 10,000,  in april, the number of cancelled flights had dropped to just above 10,000 flights.

## Multivariate Exploration

> In This section i'll be exploring how each departure delay affect departue time in the first four months as well as the relationship between fliht distance during the week to cancellation code

```python
Cancelled_flights.head()
```

```python
Cancelled_flights['CancellationCode'].unique()
```

```python
Cancel_marker= [['A','o'],
                ['B','s'],
                ['C','^'],
                ['D','X']]

for Cancel, marker in Cancel_marker:
    df_cancelled = Cancelled_flights[Cancelled_flights['CancellationCode']==Cancel]
    sb.regplot(data = df_cancelled, x = 'DayOfWeek', y = 'Distance', x_jitter = 0.4,fit_reg = False, marker=marker)
plt.legend(['A','B','C','D'],bbox_to_anchor = (1,0.5), loc = 6)
plt.legend( bbox_to_anchor = (1,0.5), loc = 6)
```

> With flights that were cancelled. we can tell that C(NAS) reasons seem to affect the cancellation of short distance flights more(1000 km or less) through the week.B(Weather) and A(Carrier) affect long distance flights with A (carrier) being the most common reason for the cancellatio of a long distance flight on days (2,3,5,6 and 7).

```python
ax = sb.pointplot(data = Delayed_flights, x = 'Month', y= 'Delfeature_time', hue = 'Delfeature',
                  dodge = 0.3, linestyles = "")
plt.legend( bbox_to_anchor = (1,0.5), loc = 6)
plt.title('Effect of each delay feature by month')
```

```python
ax=sb.barplot(data =Delayed_flights, x = 'Month', y = 'Delfeature_time', hue = 'Delfeature')
ax.legend(loc = 8 , ncol = 3, framealpha =1, title = 'Delay Feature')
plt.legend( bbox_to_anchor = (1,0.5), loc = 6)
plt.title('Effect of each delay feature by month')
```

> With the two graphs above  we are interested in looking at the effect of each delay feature by delay time across the four months. interestingly late aircraft has caused longer delay times in the first two months closely followed by weather conditions. However weather conditions cause longer delay times in the 3rd and 4th month followed by late aircraft. The other delay features are relatively consistent in terms of delay times , carrier delay (35 minutes),NAS delay (between 20 to 30 minutes), security(betwen 10 to 20 minutes)
